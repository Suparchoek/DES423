```python
import pandas as pd
```

# Step 1. â€“ Separating the features and the label

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DES423-Class3/possum.csv")
df.sample(5, random_state=44)
```

Remove any rows with missing data with.

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

Remove the unnecessary columns, then store the features and the label data in separate variables.

```python
X = df.drop(["case", "site", "Pop", "sex"], axis=1)
y = df["sex"]
```

# Step 2. â€“ Training our random forest model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=44)
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, max_features="auto", random_state=44)
rf_model.fit(X_train, y_train)
```

# Step 3. â€“ Making predictions with our model

```python
predictions = rf_model.predict(X_test)
predictions
```

```python
rf_model.predict_proba(X_test)
```

```python
importances = rf_model.feature_importances_
columns = X.columns
i = 0
while i < len(columns):
  print(f" The importance of feature' {columns[i]}' is {round (importances[i] * 100, 2)}%.")
  i += 1
```
